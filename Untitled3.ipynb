```python
# 1. Upload file Excel dari lokal ke Colab
from google.colab import files
uploaded = files.upload()

# 2. Membaca file yang diunggah
import io
import pandas as pd

# Ambil nama file yang diupload
file_name = list(uploaded.keys())[0]

# Baca Excel dari objek upload
df = pd.read_excel(io.BytesIO(uploaded[file_name]))

# 3. Tampilkan info awal
print(df.head())
print(df.info())
```

```python
# 4. Import visualisasi
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# 5. Bar Plot: 10 Prodi dengan peminat terbanyak
top10_peminat = df.nlargest(10, 'Jumlah Peminat (2024)')
plt.figure(figsize=(12, 6))
sns.barplot(x='Jumlah Peminat (2024)', y='Nama Prodi', data=top10_peminat, palette='viridis')
plt.title('10 Program Studi dengan Jumlah Peminat Tertinggi (2024)')
plt.xlabel('Jumlah Peminat')
plt.ylabel('Nama Prodi')
plt.tight_layout()
plt.show()
```

```python

# 6. Pie Chart: Distribusi Akreditasi
akreditasi_counts = df['Akreditasi'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(akreditasi_counts, labels=akreditasi_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Distribusi Akreditasi')
plt.axis('equal')
plt.show()


```

```python
# 7. Boxplot Gaji vs Proyeksi Lapangan Kerja
plt.figure(figsize=(10, 6))
sns.boxplot(x='Proyeksi Lapangan Kerja', y='Rata-Rata Gaji', data=df, palette='Set2')
plt.title('Distribusi Gaji Berdasarkan Proyeksi Lapangan Kerja')
plt.xlabel('Proyeksi Lapangan Kerja')
plt.ylabel('Rata-Rata Gaji')
plt.tight_layout()
plt.show()

```

```python
# 8. Scatter Plot Jumlah Peminat vs Daya Tampung
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Jumlah Peminat (2024)', y='Daya Tampung (2024)', hue='Akreditasi', data=df, palette='Dark2', alpha=0.7)
plt.title('Jumlah Peminat vs Daya Tampung (2024)')
plt.xlabel('Jumlah Peminat')
plt.ylabel('Daya Tampung')
plt.legend(title='Akreditasi')
plt.tight_layout()
plt.show()

```

```python
# 9. Bar Plot: Rata-rata Gaji tertinggi berdasarkan PTN
top_ptn_gaji = df.groupby('PTN')['Rata-Rata Gaji'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_ptn_gaji.values, y=top_ptn_gaji.index, palette='mako')
plt.title('10 PTN dengan Rata-Rata Gaji Tertinggi')
plt.xlabel('Rata-Rata Gaji')
plt.ylabel('PTN')
plt.tight_layout()
plt.show()
```
